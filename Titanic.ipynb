***Заполнение пропусков***

```python
#Заполнение порта посадки
df.Embarked.fillna(df.Embarked.mode()[0], inplace = True)
```

**Так как больше всего пассажиров садились в Саутгемптоне, то оставшиеся 2 пропуска решил заполнить именно им (вероятность посадки именно оттуда выше)**

```python
#Заполнение возраста
df.Age.fillna(df.Age.mean(), inplace = True)
```

```python
#Заполнение стоимости билета
df.Fare.fillna(df.Fare.mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
#Извлечение титулов и создание отдельной колонки под них
df['Title'] = df.Name.apply(lambda name: name.split(',')[1].split('.')[0].strip())
df.head()
```

```python
#Таблица количества титулов по полу
pd.crosstab(df['Title'], df['Sex'])
```

**Исходя из таблицы с количеством пассажиров с титулами можно увидеть, что некоторые титулы редко используются. Их можно объединить в какой-нибудь один, нейтральный титул**

```python
df['Title'] = df['Title'].replace(['Capt', 'Col', 'Don', 'Dr', 'Jonkheer', 'Lady', 'Major', 'Rev', 'Sir', 'the Countess'], 'Another')
df['Title'] = df['Title'].replace(['Mlle', 'Ms'], 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')

df[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

```python
#Изменяем титулы на числовые значения
title = {"Another": 1, "Master": 2, "Miss": 3, "Mr": 4, "Mrs": 5}
df['Title'] = df['Title'].map(title)
df['Title'] = df['Title'].fillna(0)
df.head()
```

```python
#Изменяем пол на числовые значения
sex = {'female': 0, 'male': 1}
df['Sex'] = df['Sex'].map(sex)
df.head()
```

```python
#Изменяем место посадки на числовые значения
embarked = {'S': 0, 'C': 1, 'Q': 2}
df['Embarked'] = df['Embarked'].map(embarked)
df.head()
```

**Убираем ненужные признаки, которые не будут использоваться в прогнозировании**

```python
df = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin'], axis=1)
```

**Перейдём к моделированию данных**

```python
X_train = df.drop('Survived', axis=1)
y_train = df['Survived']
X_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout

def create_model(optimizer='adam', loss="binary_crossentropy"):
  model = Sequential()
  model.add(Dense(32, activation='relu', input_shape=(X_train.shape[1],)))
  model.add(Dropout(0.1))
  model.add(Dense(16, activation='relu'))
  model.add(Dropout(0.1))
  model.add(Dense(8, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
  return model
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
model = create_model()
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy:.2f}')
```

**Logistic Regression** - это модель, в которой зависимая величиная является категориальной и бинарной. Т.е. применяет она только два значения: "0" и "1" (успешно и неуспешно).

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred_log_reg = clf.predict(X_test)
acc_log_reg = round(clf.score(X_train, y_train) * 100, 2)
print (str(acc_log_reg) + ' процентов')
```

**Support Vector Machine (SVM)** - это модель обучения с учителем. Она представляет данные, в виде точек на области, рассатавленных таким образом, что значения из разных категорий явно разделены между собой промежутком как можно наиболее широким. Алгоритм этого разделения называется - SVC(support vector clustering).

```python
clf = SVC()
clf.fit(X_train, y_train)
y_pred_svc = clf.predict(X_test)
acc_svc = round(clf.score(X_train, y_train) * 100, 2)
print(str(acc_svc) + ' процентов')
```

**Линейный SVM** - та же модель, но с линейным ядром.

```python
clf = LinearSVC()
clf.fit(X_train, y_train)
y_pred_linear_svc = clf.predict(X_test)
acc_linear_svc = round(clf.score(X_train, y_train) * 100, 2)
print(str(acc_linear_svc) + ' процентов')
```

**k - ближайших соседей** - алгоритм, в котором входные данные состоят из  k
  ближайших точек в пространстве признаков.

```python
clf = KNeighborsClassifier(n_neighbors = 3)
clf.fit(X_train, y_train)
y_pred_knn = clf.predict(X_test)
acc_knn = round(clf.score(X_train, y_train) * 100, 2)
print(str(acc_knn) + ' процентов')
```

**Decision tree** - структура, похожая на последовательность вычислений в которой каждый вычисляющий узел содержит некую проверку/сравнение/тест какого-либо атрибута или признака, каждая ветвь представляет результат теста, а каждый лист представляет ответ.

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred_decision_tree = clf.predict(X_test)
acc_decision_tree = round(clf.score(X_train, y_train) * 100, 2)
print(str(acc_decision_tree) + ' процентов')
```

**Random Forest** - "лес случайных решений", который выполняется за счет создания множества деревьев во время обучения.

```python
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
y_pred_random_forest = clf.predict(X_test)
acc_random_forest = round(clf.score(X_train, y_train) * 100, 2)
print(str(acc_random_forest) + ' процентов')
```

**Gaussian Naive Bayes** - алгоритм, в основе которого режит применение теоремы Байеса. Теорема Байеса описывает вероятность события, основываясь на априорном знании условий, которые могут быть связаны с событием.

```python
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred_gnb = clf.predict(X_test)
acc_gnb = round(clf.score(X_train, y_train) * 100, 2)
print(str(acc_gnb) + ' процентов')
```

**Perceptron** - это вид алгоритм классификации, который дает прогноз, основываясь на линейной функции предсказания, комбинирующей множество весов с вектором признаков.

```python
clf = Perceptron(max_iter=5, tol=None)
clf.fit(X_train, y_train)
y_pred_perceptron = clf.predict(X_test)
acc_perceptron = round(clf.score(X_train, y_train) * 100, 2)
print(str(acc_perceptron) + ' процентов')
```

**SGD(Стохастический градиентный спуск)** - метод оптимизации по минимизации целевой функции, которая записывается как сумма дифференцируемых функций. Другими словами, SGD пытается итеративно найти минимумы/максимумы.

```python
clf = SGDClassifier(max_iter=5, tol=None)
clf.fit(X_train, y_train)
y_pred_sgd = clf.predict(X_test)
acc_sgd = round(clf.score(X_train, y_train) * 100, 2)
print(str(acc_sgd) + ' процентов')
```

**Confusion Matrix** у Random Forest

```python
from sklearn.metrics import confusion_matrix
import itertools

clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
y_pred_random_forest_training_set = clf.predict(X_train)
acc_random_forest = round(clf.score(X_train, y_train) * 100, 2)
print ("Точность: %i %% \n"%acc_random_forest)
class_names = ['Survived', 'Not Survived']

cnf_matrix = confusion_matrix(y_train, y_pred_random_forest_training_set)
cnf_matrix_percent = cnf_matrix.astype('float') / cnf_matrix.sum(axis=1)[:, np.newaxis]

true_class_names = ['True Survived', 'True Not Survived']
predicted_class_names = ['Predicted Survived', 'Predicted Not Survived']

df_cnf_matrix = pd.DataFrame(cnf_matrix,
                             index = true_class_names,
                             columns = predicted_class_names)

df_cnf_matrix_percent = pd.DataFrame(cnf_matrix_percent,
                                     index = true_class_names,
                                     columns = predicted_class_names)

plt.figure(figsize = (15,5))
ax1 = plt.subplot(121)
sns.heatmap(df_cnf_matrix, annot=True, fmt='d')
ax1.title.set_text('Random Forest: значения')
ax2 = plt.subplot(122)
sns.heatmap(df_cnf_matrix_percent, annot=True)
ax2.title.set_text('Random Forest: %')
```

**Classification Report** у Random Forest

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_random_forest))
```

**Confusion Matrix** у Perceptron

```python
clf = Perceptron(max_iter=5, tol=None)
clf.fit(X_train, y_train)
y_pred_perceptron = clf.predict(X_train)
acc_perceptron = round(clf.score(X_train, y_train) * 100, 2)
print ("Точность: %i %% \n"%acc_perceptron)

cnf_matrix = confusion_matrix(y_train, y_pred_perceptron)
cnf_matrix_percent = cnf_matrix.astype('float') / cnf_matrix.sum(axis=1)[:, np.newaxis]

df_cnf_matrix = pd.DataFrame(cnf_matrix,
                             index = true_class_names,
                             columns = predicted_class_names)

df_cnf_matrix_percent = pd.DataFrame(cnf_matrix_percent,
                                     index = true_class_names,
                                     columns = predicted_class_names)

plt.figure(figsize = (15,5))
ax1 = plt.subplot(121)
sns.heatmap(df_cnf_matrix, annot=True, fmt='d')
ax1.title.set_text('Perceptron: значения')
ax2 = plt.subplot(122)
sns.heatmap(df_cnf_matrix_percent, annot=True)
ax2.title.set_text('Perceptron: %')
```
